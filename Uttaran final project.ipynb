After this we define 3 very important functions: get_genre, get_p_cast and get_cast. 
The get_genre function gives us the genres of each movie
The get_p_cast function gives us the principal cast of each movie
The get_cast function gives us the cast of each movie
    

```python
def get_genre(detail_link): 
    detail_page = requests.get(detail_link, headers = headers)
    detail_soup = BeautifulSoup(detail_page.content, "html.parser")
    ge_tr = detail_soup.find('tr', class_ = 'genres')
    ge_td = ge_tr.find('td', class_ = 'data')
    gen = ge_td.text
    gen = gen.replace(" ","")
    gen = gen.replace("\n","")
    genres = gen.split(',')
    return genres
    
```

```python
def movie_check(movie_name): #this function takes in movie name and returns the actors in it & return what
    import re #for using case sensitivity function ignorecase
    match=0
    print('\n')
    
    for i in all_movie:
        for key1, value1 in i.items():
            if (re.match(key1,movie_name,re.IGNORECASE)): #checking the user input with the case sensitivity requirement
                match=1
                print("The actors/actresses in this movie are : ")
                
                for key2,value2 in value1.items(): #this loop iterates through the dict to check and print the relevant key(actors) and values(genres)
                    for j in range(0,len(key2)):
                        print(key2[j])
                    print('\n')
                    print("The genres of the movie are : ")
                    for k in range(0,len(value2)):
                        print(value2[k])
            
    if(match==0):
        print("Sorry the name was not found")
        main_menu()
                
    
```

```python
def actor_details(actor_name): #this function on entering name of actor returns his movies and genres and their count
    import re
    genre_dict={} #dict that will eventually store all actor details like movie name and genres 
    match=0
    serial=1
    flag=0
    
    for i in all_movie: 
        for key1, value1 in i.items():#first layer of the dictionary
            for key2,value2 in value1.items(): #accessing the second layer of the all_movie dictionary
                    for j in range(0,len(key2)):
                        if(re.match(key2[j],actor_name,re.IGNORECASE)):
                            match=1
                            if flag==0:
                                print("The names of the movies that "+ actor_name + " has acted in are : ")
                                flag=1
                            print(str(serial) + ". " + key1) #key1 will display the name of the movie
                            serial+=1
                            for k in range(0,len(value2)):
                                if(value2[k] in genre_dict):
                                    genre_dict[value2[k]]=genre_dict[value2[k]]+1 #updating the count of the genres 
                                else: 
                                    genre_dict[value2[k]]=1 #in case if the genre did not exist before
                            
    print('\n')
    print("The total types of genres and their frequencies for this person are as follows --")
    for key, value in genre_dict.items():
        print(key, ' : ', value)
    print('\n')
    if(match==0):
        print("Sorry the name was not found!! Please try again")
        main_menu()
    
```

```python
def comparison():
    from numpy import dot
    from numpy.linalg import norm
    import re
    print("Enter the names of two persons you want to compare")
    actor_name1=input("Enter name of the first actor/actress : ")
    print(actor_details(actor_name1))
    actor_name2=input("Enter name of the second actor/actress : ")
    print(actor_details(actor_name2))
    actor1_genres = [] # first we create an empty list to hold all the genres of actor1
    for i in all_movie:
        for key, value in i.items():  #this breaks each movie dictionary into movie name (key), and sub dictionary(value)
            for subkey, subvalue in value.items():  #this breaks the subdictionary (value) into actor (sub key), and genres (sub value)
                if actor_name1 in subkey:  # subkey is a tuple of actors, so we check if the selected actor is in there
                    actor1_genres.extend(subvalue) #subvalue is a list of genres, so we use extend


#repeating the same process for actor2
    actor2_genres = [] 
    for i in all_movie:
        for key, value in i.items():  
            for subkey, subvalue in value.items(): 
                if actor_name2 in subkey: 
                    actor2_genres.extend(subvalue)
            
    set_actor1 = set(actor1_genres) #the set function would take care of duplicates
    set_actor2 = set(actor2_genres)
    combine_set = set_actor1.union(set_actor2) #union will also remove all duplicates after combining

    actor1_dict = dict()
    for i in combine_set:
        actor1_dict[i] = 0

    actor2_dict = dict()
    for i in combine_set:
        actor2_dict[i] = 0
    
    for i in actor1_genres:
        actor1_dict[i] = actor1_dict[i] + 1
    
#repeating the same process for actor2
    for i in actor2_genres:
        actor2_dict[i] = actor2_dict[i] + 1
    
    actor1_vector = list(actor1_dict.values())  # converting into list of vector values
    actor2_vector = list(actor2_dict.values())  # converting into list of vector values
    
    cosine_sim = dot(actor1_vector, actor2_vector)/(norm(actor1_vector)*norm(actor2_vector)) #calc the cosine similarity
    
    print('\n')
    print("The cosine similarity score between them is ",end="")
    print(cosine_sim)
```
